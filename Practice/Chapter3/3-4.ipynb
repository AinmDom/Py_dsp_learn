在乐理中，“滑奏”（glissando）是指音符从一个音高滑向另一个，与啁啾类似；寻找一个滑奏的录音并绘制其片段的频谱。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import librosa

# 设置中文字体
plt.rcParams["font.family"] = ["SimHei", "Times New Roman"]  # 使用系统中存在的字体

sg, fs = librosa.load(r"D:\course\CUC_0\Py_dsp_learn\Sample_sound\violin-glissando-ascending-a-h4n.wav")
```

```python
# 时频谱spectrogram
S = librosa.feature.melspectrogram(y=sg,sr=2*fs, n_fft=int(0.1*fs), n_mels=512)
# 转换为 dB 单位
S_dB = librosa.power_to_db(S, ref=np.max)

# 绘制梅尔频谱图
plt.figure(figsize=(18,6))
librosa.display.specshow(S_dB,sr=fs,x_axis='time',y_axis='mel')  # y 轴为梅尔频率
plt.colorbar(label='dB')
plt.ylim(512, fs/2)
plt.title('梅尔频谱图')
plt.tight_layout()
plt.show()
```

绘制频谱后，可以明显看到滑奏的谐波结构，但melspectrogram()的频率与时间分辨率似乎不太好调整。使用Audition查看的频谱分辨率要好很多。
