A Soft Murmur是一个播放自然源噪声混合的网站，包括雨声、波浪和风声等，从这里下载几个文件，计算并绘制各个信号的频谱，它们的功率谱和白噪声、粉噪或者布朗噪声类似吗？

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import librosa
from IPython.display import Audio

# 设置中文字体
plt.rcParams["font.family"] = ["SimHei", "Times New Roman"]  # 使用系统中存在的字体
```

```python
# sg, fs = librosa.load(r"D:\course\CUC_0\Py_dsp_learn\Sample_sound\rain-from-inside.wav")  # 雨声
sg, fs = librosa.load(r"D:\course\CUC_0\Py_dsp_learn\Sample_sound\seashore-wave-1.wav")  # 海浪声
```

```python
# 频谱
n_fft = 1024  # FFT窗口大小
sn_spec = np.fft.rfft(sg, n=n_fft)  # 加直流偏离，对应频谱上f=0的值
amplitude = np.abs(sn_spec)  # 幅度谱
frequency = np.fft.rfftfreq(n_fft, d=1 / fs)  # 频率轴

# 绘制幅度谱
plt.figure(figsize=(18, 6))
plt.plot(frequency, amplitude)
plt.title('原幅度谱（FFT）')
plt.xlabel('频率 (Hz)')
plt.ylabel('幅度')
plt.grid(True)
# plt.xlim(200, 700)
plt.show()
Audio(sg, rate=fs)
```

雨声与海浪声整体上有粉红噪声的特点，但低频部分一般更少。
