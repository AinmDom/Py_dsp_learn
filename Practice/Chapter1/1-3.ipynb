```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from IPython.display import Audio

duration = 2.0       # 信号持续时间（秒）
sample_rate = 44100  # 采样率（Hz）
base_freq = 440      # 正弦波频率（Hz，对应 A 音）
amplitude = 0.5      # 振幅

# 生成时间序列
t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)

# 生成正弦信号
sin_signal = amplitude * np.sin(2 * np.pi * base_freq * t)
cos_signal = amplitude * np.cos(2 * np.pi * base_freq * t)

sin_signal_2h = amplitude * np.sin(2 * np.pi * 666 * t)
```

```python
y = sin_signal + cos_signal + sin_signal_2h
Audio(y, rate=sample_rate)
```

当相加信号非谐波关系时，能明显感到不相关的音轨之间的分离感；反之则很难察觉
